```python
def r2_t0_XP(name_of_csv,your_title):
    
    read = pd.read_csv(f'{name_of_csv}')
    t0 = read[["t0"]]
    #t0 = t0 - ???
    inc = read[["inclination"]]
    r2 = read[["r_squared"]]
    mass = read[["mass"]]
    chi = read[["chi"]]
    #print(mass)
    #print(chi)
    
    #slicing times
    t0_0 = t0.iloc[:13, :]
    t0_pi6 = t0.iloc[13:26, :]
    t0_pi3 = t0.iloc[26:39, :]
    t0_pi2 = t0.iloc[39:65, :]
    
    #slicing inclination
    inc_0 = inc.iloc[:13, :]
    inc_pi6 = inc.iloc[13:26, :]
    inc_pi3 = inc.iloc[26:39, :]
    inc_pi2 = inc.iloc[39:65, :]

    #slicing r squared
    r2_0 = r2.iloc[:13, :]
    r2_pi6 = r2.iloc[13:26, :]
    r2_pi3 = r2.iloc[26:39, :]
    r2_pi2 = r2.iloc[39:65, :]
    
    plt.figure(figsize=(10,6))
    plt.scatter(t0_0,(1-(r2_0)), marker='^',s=100,label='0', color='deeppink')
    plt.scatter(t0_pi6,(1-(r2_pi6)), marker='*',s=150,label='pi/6', color='mediumturquoise')
    plt.scatter(t0_pi3,(1-(r2_pi3)), marker='.',s=150,label='pi/3', color='rebeccapurple')
    plt.scatter(t0_pi2,(1-(r2_pi2)), marker='s',s=100,label='pi/2', color='sandybrown')
    plt.legend(title='Inclination',fontsize=15)
    plt.ylabel('$1-R^2$',fontsize=20)
    plt.yscale('log')
    #plt.xlim(0.1,1.1)
    plt.xlabel('t0',fontsize=20)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.title(f'{your_title}',fontsize=25)
    plt.show()
```

```python
r2_t0_XP('XP_eqM_noX_2ot.csv','IMRPhenomXP ($M1=M2$) ($\chi=0$)')
```

```python
r2_t0_XP('XP_eqM_add_spin_2ot.csv','IMRPhenomXP ($M1=M2$) ($\chi=0.5$)')
```
