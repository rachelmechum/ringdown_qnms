```python
from pycbc.waveform import get_td_waveform
import matplotlib.pyplot as plt
from pycbc.types import TimeSeries
import numpy as np
```

```python
hp, hc = get_td_waveform(approximant="IMRPhenomXPHM",
                         mass1=40,
                         mass2=40,
                         delta_t=1.0/16384,
                         f_lower=30,
                         distance=1e-14,
                         inclination=0)

# sample times ---> timeseries
hp_smp_ts = TimeSeries(hp.sample_times, delta_t=1/16384, epoch=0)
# using crop on timeseries sample times - use cond for cropped sample times (aka ringdown analysis)
cond = hp_smp_ts.crop(abs(hp.start_time),0)                                      

# h+/hx data ---> timeseries
hp_data = TimeSeries(hp, delta_t=1/16384, epoch=0)  
# using crop on timeseries data (use cond_hp/hc for ringdown anaylsis)
cond_hp = hp_data.crop(abs(hp.start_time),0)
# turning into ndarray
hp_signal = cond_hp.numpy()                  

# repeat
hc_data = TimeSeries(hc, delta_t=1/16384, epoch=0)              
cond_hc = hc_data.crop(abs(hp.start_time),0)
hc_signal = cond_hc.numpy()                                         

# use for hp+hc ringdown analysis
time = cond.numpy()              


#plt.figure(figsize=pylab.figaspect(0.4))
fig1, ax1 = plt.subplots(figsize=(10,5))
fig1.patch.set_facecolor('xkcd:black')
ax1.set_facecolor('xkcd:black')
ax1.spines['bottom'].set_color('white')
ax1.spines['left'].set_color('white')
ax1.spines['top'].set_color('white')
ax1.spines['right'].set_color('white')
ax1.tick_params(axis='both',labelsize=20,colors='white')
#plt.grid()
plt.plot(cond, cond_hc, label='IMR XPHM Cross',color='mediumturquoise')
plt.xlim(0, 0.03)
plt.title('IMRPhenomXPHM Ringdown',fontsize=25,color='white')
plt.xlabel('Time (s)',fontsize=20,color='white')
plt.ylabel('Strain',fontsize=20,color='white')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
```

```python
def mode_change(l,m):
    hp, hc = get_td_waveform(approximant="IMRPhenomXPHM",
                             mass1=40,
                             mass2=39.99,
                             delta_t=1.0/16384,
                             f_lower=30,
                             distance=1e-14,
                             inclination=0,
                             mode_array=[[l,m]])

    # sample times ---> timeseries
    hp_smp_ts = TimeSeries(hp.sample_times, delta_t=1/16384, epoch=0)
    # using crop on timeseries sample times - use cond for cropped sample times (aka ringdown analysis)
    cond = hp_smp_ts.crop(abs(hp.start_time),0)                                      

    # h+/hx data ---> timeseries
    hp_data = TimeSeries(hp, delta_t=1/16384, epoch=0)  
    # using crop on timeseries data (use cond_hp/hc for ringdown anaylsis)
    cond_hp = hp_data.crop(abs(hp.start_time),0)
    # turning into ndarray
    hp_signal = cond_hp.numpy()                  

    # repeat
    hc_data = TimeSeries(hc, delta_t=1/16384, epoch=0)              
    cond_hc = hc_data.crop(abs(hp.start_time),0)
    hc_signal = cond_hc.numpy()                                         

    # use for hp+hc ringdown analysis
    time = cond.numpy()              


    #plt.figure(figsize=pylab.figaspect(0.4))
    fig1, ax1 = plt.subplots(figsize=(10,5))
    fig1.patch.set_facecolor('xkcd:black')
    ax1.set_facecolor('xkcd:black')
    ax1.spines['bottom'].set_color('white')
    ax1.spines['left'].set_color('white')
    ax1.spines['top'].set_color('white')
    ax1.spines['right'].set_color('white')
    ax1.tick_params(axis='both',labelsize=20,colors='white')
    #plt.grid()
    plt.plot(cond, cond_hc, label=f'{l},{m}',color='mediumturquoise')
    plt.xlim(0, 0.03)
    plt.xlabel('Time (s)',fontsize=20,color='white')
    plt.ylabel('Strain',fontsize=20,color='white')
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
```

```python
mode_change(2,2)
```

```python
mode_change(3,3)
```

```python
mode_change(4,4)
```
