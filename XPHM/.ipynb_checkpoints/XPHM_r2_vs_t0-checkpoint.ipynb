```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def r2_t0_XPHM(name_of_csv,your_title):
    
    read = pd.read_csv(f'{name_of_csv}')
    t0 = read[["t0"]]
    t0 = t0 - 3.8
    inc = read[["inclination"]]
    r2 = read[["r_squared"]]
    mass = read[["mass"]]
    chi = read[["chi"]]
    #print(mass)
    #print(chi)
    
    #slicing times
    t0_0 = t0.iloc[:12, :]
    t0_pi6 = t0.iloc[12:24, :]
    t0_pi3 = t0.iloc[24:36, :]
    t0_pi2 = t0.iloc[36:48, :]

    #slicing inclination
    inc_0 = inc.iloc[:12, :]
    inc_pi6 = inc.iloc[12:24, :]
    inc_pi3 = inc.iloc[24:36, :]
    inc_pi2 = inc.iloc[36:48, :]

    #slicing r squared
    r2_0 = r2.iloc[:12, :]
    r2_pi6 = r2.iloc[12:24, :]
    r2_pi3 = r2.iloc[24:36, :]
    r2_pi2 = r2.iloc[36:48, :]
    
    plt.figure(figsize=(10,6))
    plt.scatter(t0_0,(1-(r2_0)), marker='^',s=100,label='0', color='deeppink')
    plt.scatter(t0_pi6,(1-(r2_pi6)), marker='*',s=150,label='pi/6', color='mediumturquoise')
    plt.scatter(t0_pi3,(1-(r2_pi3)), marker='.',s=150,label='pi/3', color='rebeccapurple')
    plt.scatter(t0_pi2,(1-(r2_pi2)), marker='s',s=100,label='pi/2', color='sandybrown')
    plt.legend(title='Inclination',fontsize=15)
    plt.ylabel('$1-R^2$',fontsize=20)
    plt.yscale('log')
    #plt.ylim(0.1,1.1)
    plt.xlabel('t0',fontsize=20)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.title(f'{your_title}',fontsize=25)
    plt.show()
```

```python
#this is based on old code...only use if runs don't finish in time
r2_t0_XPHM('allmodes_n012ot.csv','21 22 32 33 44 (n=0,1,2)')
```
